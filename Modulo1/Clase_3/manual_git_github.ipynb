## 1. ¿Qué es Git y por qué existe?

¿Alguna vez tuviste una carpeta así?

```
script_v1.py
script_v2.py
script_final.py
script_final_DEFINITIVO.py
script_final_DEFINITIVO_2.py
```

Eso es **control de versiones manual**. Funciona hasta que hay 10 archivos, 3 personas trabajando y nadie sabe cuál es el bueno.

**Git** es un sistema de control de versiones distribuido. Resuelve tres problemas:

| Problema | Cómo lo resuelve Git |
|----------|---------------------|
| "¿Qué cambié y cuándo?" | Historial limpio: cada cambio registrado con autor, fecha y descripción |
| "Estamos pisándonos el código" | Trabajo paralelo en ramas independientes |
| "Rompí algo y no sé cómo volver" | Reversibilidad total: puedes volver a cualquier punto anterior |

### Git ≠ GitHub

| | Git | GitHub |
|---|---|---|
| ¿Qué es? | Herramienta que vive en tu computadora | Plataforma en la nube para compartir repositorios |
| ¿Puede existir sin el otro? | ✅ Sí | ❌ No (necesita Git) |
| Analogía | El sistema de guardado | La nube donde guardas las fotos |

> **Git** gestiona el historial local. **GitHub** lo publica y permite la colaboración.

## 2. Instalación y configuración inicial

### Verificar si Git está instalado
# Ejecutar en terminal
git --version
# Esperado: git version 2.x.x
**Si no tienes Git instalado (Windows):**
1. Descargar desde: https://git-scm.com/download/win
2. Durante la instalación: dejar **todas las opciones por defecto**
3. Opción importante: *"Use Git from the command line and also from 3rd-party software"*

**Mac:**
```bash
brew install git
```

### Configurar tu identidad (solo una vez por computadora)
# Decirle a Git quién eres
# Cada commit que hagas llevará este nombre y email
git config --global user.name "Tu Nombre"
git config --global user.email "tuemail@ejemplo.com"

# Verificar que quedó guardado
git config --list
> **`--global`** significa que esta config aplica a todos tus repos en esta computadora.  
> Si en un proyecto específico necesitas un email distinto, ejecuta el mismo comando **sin** `--global` dentro de esa carpeta.

## 3. Crear un repositorio local

Un **repositorio** es simplemente una carpeta que Git está vigilando. Nada más.
# Crear la carpeta del proyecto y entrar
mkdir mi-proyecto
cd mi-proyecto

# Inicializar el repositorio
git init
# Verificar que se creó la carpeta oculta .git
ls -a       # Mac / Linux
dir /a      # Windows
La carpeta `.git` contiene todo el historial.  
⚠️ **No la toques manualmente.** Si la borras, pierdes el historial (no el código, pero sí todos los registros de cambio).

### Abrir el proyecto en VS Code
# Abre VS Code en la carpeta actual
# Vas a usar esto constantemente
code .

## 4. El ciclo fundamental: add → commit → log

Este es el corazón de Git. Todo lo demás es una variación de esto.

```
[Tu carpeta]  →  git add  →  [Área de staging]  →  git commit  →  [Historial]
 (Working Dir)    "Seleccionas qué va"              "La foto queda guardada"
```

**Analogía:** Git es una cámara fotográfica.
- `git add` = elegir qué aparece en la foto
- `git commit` = sacar la foto
- historial = tu álbum (inmutable, siempre consultable)

### Paso a paso
# 1. Ver el estado actual — úsalo SIEMPRE antes de hacer algo
git status
# 2a. Agregar un archivo específico al staging
git add analisis.py

# 2b. Agregar TODOS los archivos modificados
git add .

# Ver el estado después del add
git status
# 3. Crear el commit con un mensaje descriptivo
git commit -m "feat: agrego función de análisis de promedios"

# Convención de mensajes (Conventional Commits):
# feat:     nueva funcionalidad
# fix:      corrección de bug
# docs:     cambios en documentación
# refactor: reestructurar sin cambiar comportamiento
# chore:    tareas de mantenimiento (ej: .gitignore, dependencias)
# 4. Ver el historial
git log                # versión completa
git log --oneline      # versión compacta — úsala siempre en el día a día

# Para salir del paginador si aparece (END): presiona  q
# Ver exactamente qué cambió entre el commit anterior y el actual
git diff HEAD~1 HEAD
# Líneas en verde = agregadas | Líneas en rojo = eliminadas

## 5. .gitignore — qué no debe entrar al repositorio

Hay archivos que Git **no debería rastrear jamás:**
- Contraseñas y API keys (`.env`)
- Archivos temporales de Python (`__pycache__`, `.pyc`)
- Entornos virtuales (`venv/`, `.venv/`)
- Configuraciones del editor (`.vscode/`, `.idea/`)
# Crear el archivo .gitignore desde terminal
echo __pycache__/ > .gitignore
echo .env >> .gitignore
echo *.pyc >> .gitignore
echo venv/ >> .gitignore
echo .vscode/ >> .gitignore

# Commitearlo
git add .gitignore
git commit -m "chore: agrego .gitignore"
> **Regla de oro:** el archivo `.env` con credenciales **nunca** debe llegar a GitHub. Agréganlo al `.gitignore` antes del primer commit.

Puedes usar plantillas predefinidas en: https://github.com/github/gitignore

## 6. Conectar con GitHub (remoto)

Hasta ahora todo vive solo en tu computadora. GitHub es el respaldo y el espacio de colaboración.

### Crear el repositorio en GitHub

1. Ir a https://github.com → botón **New repository**
2. Nombre del repo (ej: `mi-proyecto`)
3. Visibilidad: **Public** (para construir portafolio)
4. ⚠️ **NO** marcar "Initialize with README" si ya tienes commits locales
5. Copiar la URL que te da GitHub

### Conectar tu repo local con GitHub
# Conectar el repo local con el remoto
git remote add origin https://github.com/tu-usuario/mi-proyecto.git

# Verificar la conexión
git remote -v

# Subir todo al remoto por primera vez
# El flag -u crea el vínculo permanente entre tu rama local y la remota
git push -u origin main
# A partir de ahora, para subir cambios simplemente:
git push

# Para traer cambios que otros subieron:
git pull
### El flujo completo de trabajo diario

```
Modifico código  →  git add  →  git commit  →  git push
                                                   ↑
                                              llega a GitHub
```

### Agregar un README
# El README es lo primero que ve cualquier persona que entra a tu repo
# Es tu carta de presentación
echo "# Mi Proyecto" > README.md
echo "Descripción breve del proyecto." >> README.md
o lo puedes crear directamente en vs code, de esa forma lo hicimos en la clase

git add README.md
git commit -m "docs: agrego README inicial"
git push

## 7. Ramas: trabajar sin miedo

### ¿Para qué sirven las ramas?

Imagina que `main` es tu código en producción. Un cliente lo está usando ahora mismo. Necesitas agregar una funcionalidad nueva. ¿La haces directamente en `main`? **No.** Creas una rama, trabajas ahí, y cuando está lista y probada, la fusionas.

```
main:              A --- B --- C ----------------------------- M  (merge)
                               \                             /
feature/nueva-func:             D --- E --- F --- G ---------
```

`main` siguió su vida. La nueva función se desarrolló en paralelo. Cuando estuvo lista, se fusionó.

### Comandos de ramas
# Ver todas las ramas (el * indica en cuál estás)
git branch
# Crear una rama nueva Y moverse a ella (forma moderna, Git 2.23+)
git switch -c feature/nueva-funcion

# Forma clásica (también válida)
git checkout -b feature/nueva-funcion
# Hacer cambios, commitearlos en la rama
git add .
git commit -m "feat: agrego nueva funcionalidad"

# Subir la rama al remoto
git push -u origin feature/nueva-funcion
# Volver a main
git switch main

# Fusionar la rama en main (merge local)
git merge feature/nueva-funcion

# Verificar que los cambios llegaron
git log --oneline

# Subir main actualizado
git push
# Opcional: eliminar la rama después del merge
git branch -d feature/nueva-funcion          # local
git push origin --delete feature/nueva-funcion  # remoto
### Convención de nombres para ramas

| Prefijo | Uso |
|---------|-----|
| `feature/` | Nueva funcionalidad |
| `fix/` | Corrección de bug |
| `docs/` | Cambios de documentación |
| `refactor/` | Reestructuración de código |
| `hotfix/` | Corrección urgente en producción |

## 8. Clone vs Fork: cuándo usar cada uno

| | **Clone** | **Fork** |
|---|---|---|
| ¿Qué hace? | Copia el repo a **tu máquina** | Copia el repo a **tu cuenta de GitHub** |
| ¿Cuándo usarlo? | Cuando **eres parte del equipo** y tienes acceso | Cuando **no tienes acceso** y quieres contribuir o experimentar |
| ¿Puedes hacer push directo? | ✅ Sí (si tienes permisos) | ❌ No — debes crear un PR desde tu fork |
| Ejemplo real | Tu trabajo en el bootcamp | Contribuir a un proyecto open source |

### Cuándo hacer clone
- Tu empresa te da acceso al repositorio del producto
- Tu instructor comparte un repo de ejercicios y todos trabajan en él
- Es tu propio repo en GitHub y quieres trabajar desde otra computadora

### Cuándo hacer fork
- Quieres usar el código de alguien como base para tu propio proyecto
- Encontraste un bug en una librería open source y quieres proponer la corrección
- Quieres experimentar con código ajeno sin afectar el original
# Clonar un repositorio (primera vez que bajas un repo a tu máquina)
git clone https://github.com/usuario/nombre-repo.git

# Entrar a la carpeta que se creó
cd nombre-repo
**Fork** se hace desde la interfaz de GitHub (botón Fork en la esquina superior derecha del repo).  
Luego clonas **tu fork** (no el original):

```bash
git clone https://github.com/TU-usuario/nombre-repo.git
```

La diferencia estará en la URL: apunta a tu cuenta, no a la cuenta original.

## 9. Flujo de trabajo colaborativo y Pull Requests

Este es el flujo que usan los equipos de desarrollo profesional todos los días.

```
1. git clone / git pull   → Tener el código actualizado
2. git switch -c rama     → Crear rama para tu tarea
3. (trabajar)             → Hacer cambios
4. git add + git commit   → Guardar en historial
5. git push               → Subir la rama a GitHub
6. Pull Request           → Pedir revisión y merge
7. Code review + merge    → Tech Lead aprueba y fusiona
8. git pull en main       → Actualizar tu local
```

### El flujo paso a paso
# PASO 1: Siempre empieza con main actualizado
git switch main
git pull
# PASO 2: Crear tu rama de trabajo
git switch -c feature/mi-tarea
# PASO 3 y 4: Trabajar, añadir, commitear
# (haces tus cambios en el código)
git add .
git commit -m "feat: descripción de lo que hice"
# PASO 5: Subir la rama
git push -u origin feature/mi-tarea
**PASO 6 — Crear el Pull Request en GitHub:**
1. Ir a tu repo en GitHub
2. GitHub detecta la rama nueva y muestra un botón: **"Compare & pull request"**
3. Escribir un título descriptivo y descripción de los cambios
4. Asignar reviewers si aplica
5. Click en **"Create pull request"**

**PASO 7 — El Tech Lead revisa:**
- Puede ver los cambios línea por línea (verde = añadido, rojo = eliminado)
- Puede comentar líneas específicas
- Aprueba con **"Approve"** o pide cambios con **"Request changes"**
- Una vez aprobado: **"Merge pull request"**
# PASO 8: Después del merge, actualizar tu main local
git switch main
git pull

## 10. Resolución de conflictos

Un **conflicto** ocurre cuando dos personas modificaron la misma línea del mismo archivo en ramas distintas. Git no sabe cuál versión conservar — eso lo decidís vos.

### Cómo se ve un conflicto en el archivo
```python
<<<<<<< HEAD
    print(f"Total: {sum(lista)}")
=======
    return sum(lista)
>>>>>>> main
```

- **`<<<<<<< HEAD`** → lo que tiene tu rama actual
- **`=======`** → separador
- **`>>>>>>> main`** → lo que viene de la otra rama

### Cómo resolverlo
# 1. Actualizar main local
git switch main
git pull

# 2. Ir a tu rama y traer los cambios de main
git switch feature/mi-rama
git merge main
# → Aquí Git avisa si hay conflictos
**3. Abrir el archivo en VS Code y editar manualmente:**
- Eliminar las líneas con `<<<<<<<`, `=======`, `>>>>>>>`
- Dejar solo el código correcto (puede ser una versión, la otra, o una combinación)
- Guardar el archivo
# 4. Commitear la resolución
git add .
git commit -m "fix: resuelvo conflicto en dataset_ventas.py"
git push
> **Tip:** VS Code tiene una interfaz visual para resolver conflictos. Abre el archivo con conflicto y verás botones: *"Accept Current Change"*, *"Accept Incoming Change"*, *"Accept Both Changes"*.

## 11. Errores frecuentes y cómo resolverlos

### ⚠️ Error 1 — Commiteé en `main` sin querer
# Solución: mover el commit a una rama nueva
git switch -c feature/mi-cambio-sin-querer   # crear rama con el commit
git switch main
git reset --soft HEAD~1   # deshacer el commit en main, conservando los cambios en staging
### ⚠️ Error 2 — Modifiqué un archivo y quiero descartar los cambios
# Antes de hacer git add (solo en working directory)
git restore archivo.py

# Si ya hice git add pero NO commit
git restore --staged archivo.py   # saca del staging
git restore archivo.py            # descarta los cambios
### ⚠️ Error 3 — El push fue rechazado

```
error: failed to push some refs to 'origin/main'
```

Pasa cuando alguien más subió cambios que tu local no tiene. **Nunca hagas force push en main.**
# Primero traer los cambios remotos, luego subir los tuyos
git pull --rebase origin main
# Resolver conflictos si los hay
git push
### ⚠️ Error 4 — git log interminable, no puedo escribir
# Si ves (END) en la terminal:
# Presiona la tecla:  q

# Mejor: usa siempre la versión compacta
git log --oneline
### ⚠️ Error 5 — "detached HEAD"

```
HEAD detached at a3f2b1c
```

Git está apuntando a un commit directamente, no a una rama. No es peligroso, pero no commitees desde ahí.
# Volver a main
git switch main
### El comando que siempre salva

Antes de hacer cualquier cosa que no estés seguro/a, corre esto:
git status
# Git te dirá exactamente dónde estás y qué tienes disponible

## 12. Referencia rápida: todos los comandos

### Configuración
git config --global user.name "Tu Nombre"
git config --global user.email "email@ejemplo.com"
git config --list                               # ver toda la config
### Repositorio local
git init                        # inicializar repo
git status                      # estado actual (úsalo siempre)
git add archivo.py              # agregar archivo al staging
git add .                       # agregar todos los archivos
git commit -m "tipo: mensaje"   # crear commit
git log --oneline               # historial compacto
git diff HEAD~1 HEAD            # qué cambió en el último commit
### Deshacer cambios
git restore archivo.py          # descartar cambios (antes de add)
git restore --staged archivo.py # sacar del staging (después de add)
git reset --soft HEAD~1         # deshacer último commit (conserva cambios)
### Ramas
git branch                      # listar ramas
git switch -c nombre-rama       # crear y moverse a rama nueva
git switch main                 # volver a main
git merge nombre-rama           # fusionar rama en la actual
git branch -d nombre-rama       # eliminar rama local
### Trabajo con remoto (GitHub)
git remote add origin URL       # conectar con GitHub
git remote -v                   # ver conexión remota
git push -u origin main         # primer push
git push                        # pushes siguientes
git pull                        # traer cambios del remoto
git pull --rebase origin main   # traer y rebasar (ante push rechazado)
git clone URL                   # clonar repositorio
# Referencia rápida — Comandos del día

```bash
# CONFIGURACIÓN
git config --global user.name "Nombre"
git config --global user.email "email@ejemplo.com"
git config --list

# REPOSITORIO LOCAL
git init                          # inicializar repo
git status                        # estado actual
git add archivo.py                # agregar archivo al staging
git add .                         # agregar todo
git commit -m "tipo: mensaje"     # crear commit
git log --oneline                 # ver historial compacto
git diff HEAD~1 HEAD              # ver cambios entre commits

# DESHACER
git restore archivo.py            # descartar cambios sin add
git restore --staged archivo.py   # sacar del staging
git reset --soft HEAD~1           # deshacer último commit (conserva cambios)

# RAMAS
git branch                        # listar ramas
git switch -c nombre-rama         # crear y cambiar a rama nueva
git switch main                   # volver a main
git merge nombre-rama             # fusionar rama en la actual

# REMOTO
git remote add origin URL         # conectar con GitHub
git push -u origin main           # primer push
git push                          # pushes siguientes
git pull                          # traer cambios del remoto
git clone URL                     # clonar repositorio
```
