## Variables y tipos de datos
Definir las variables con nombres que hagan referencia a su contenido



```python
# Variable para almacenar el nombre del estudiante
nombre = "Erika Restrepo"
edad = 22
correo = "erika@gmail.com"
activo = True
promedio = 4.5

# Imprimir la información del estudiante
print(f"Nombre: {nombre}")
print(f"Edad: {edad} -> tipo {type(edad)}")
print(f"Correo: {correo}")
print(f"Activo: {activo}")
print(f"Promedio: {promedio}")
```

```python
# Variable para almacenar el nombre del estudiante digitado por el usuario
nombre = input("Ingrese el nombre del estudiante: ")
edad = int(input("Ingrese la edad del estudiante: "))
correo = input("Ingrese el correo del estudiante: ")
activo = input("¿El estudiante está activo? (True/False): ").lower() == 'true'
promedio = float(input("Ingrese el promedio del estudiante: "))
print(f"Registrado {nombre} con edad {edad}, correo {correo}, activo: {activo} y promedio {promedio}")
```

str -> string cadena de caracteres
int -> Enteros numero enteros entre y 
float ->  Float numeros con decimales
bool -> Booleano True\False


## Listas
> Es una coleccion ordenada de elementos, se implementa como un array dinamico, su acceso lo tiene por indices.


```python
# Lista de estudiantes
estudiantes = [
    "Graciela Restrepo",
    "Erika Restrepo",
    "Carlos Mendoza",
    "Ana López"
]

# Imprimir cada estudiante de la lista
print(f"Primer estudiante: {estudiantes[0]}")  # Imprime el primer estudiante
print(f"Segundo estudiante: {estudiantes[1]}")  # Imprime el segundo estudiante
print(f"Tercer estudiante: {estudiantes[2]}")  # Imprime el tercer estudiante
print(f"Cuarto estudiante: {estudiantes[3]}")  # Imprime el cuarto estudiante

# Imprimir el estudiante en la ultima posición de la lista
print(f"Último estudiante: {estudiantes[-1]}")  # Imprime "Ana López"

# Imprimir la cantidad de estudiantes en la lista
print(f"Cantidad de estudiantes: {len(estudiantes)}")

# Agregar un nuevo estudiante a la lista
estudiantes.append("Luis Gómez")
print(f"Nuevo estudiante agregado: {estudiantes[-1]}")  # Imprime el nuevo estudiante agregado
print(f"Lista actualizada de estudiantes: {estudiantes}")

# Agregar un nuevo estudiante a la lista desde el input del usuario
nuevo_estudiante = input("Ingrese el nombre del nuevo estudiante: ")
estudiantes.append(nuevo_estudiante)
print(f"Nuevo estudiante agregado: {estudiantes[-1]}")  # Imprime el nuevo estudiante agregado
print(f"Lista actualizada de estudiantes: {estudiantes}")
```

# Diccionarios
> Son pares de clave-valor

```python
# Registro de estudiantes:
estudiante = {
    "nombre": "Erika Restrepo",
    "edad": 20,
    "correo": "Erika.Restrepo@university.edu",
    "activo": True,
    "promedio": 4.5
}

# Imprimir la información del estudiante
print(f"Nombre: {estudiante['nombre']}")
print(f"Edad: {estudiante['edad']} -> tipo {type(estudiante['edad'])}")
print(f"Correo: {estudiante['correo']}")
print(f"Activo: {estudiante['activo']} -> tipo {type(estudiante['activo'])}")
print(f"Promedio: {estudiante['promedio']} -> tipo {type(estudiante['promedio'])}")

# Imprimier el diccionario completo del estudiante
print(f"Información completa del estudiante: {estudiante}")

# Actualizar el correo del estudiante en el diccionario
estudiante["correo"] = "erika@university.edu"
print(f"Correo actualizado: {estudiante['correo']}")
```

## Listas y Diccionarios
Agregamos varios estudiantes

```python
estudiante = {
    "nombre": "Erika Restrepo",
    "edad": 20,
    "correo": "Erika.Restrepo@university.edu",
    "activo": True,
    "promedio": 4.5
}

estudiantes = [
    estudiante,
    {
        "nombre": "Camilo Restrepo",
        "edad": 20,
        "correo": "camilo@example.com",
        "activo": True,
        "promedio": 4.5
    },
    {
        "nombre": "Carlos Mendoza",
        "edad": 22,
        "correo": "carlos@example.com",
        "activo": False,
        "promedio": 3.8
    }
]

# Imprimir el nombre de cada estudiante en la lista de estudiantes
print("Estudiantes registrados:")
print(estudiantes)
```

Se puede modularizar el codigo

```python
estudiante1 = {
    "nombre": "Erika Restrepo",
    "edad": 20,
    "correo": "Erika.Restrepo@university.edu",
    "activo": True,
    "promedio": 4.5
}

estudiante2 = {
    "nombre": "Camilo Restrepo",
    "edad": 20,
    "correo": "camilo@example.com",
    "activo": True,
    "promedio": 4.5
}

estudiante3 = {
    "nombre": "Carlos Mendoza",
    "edad": 22,
    "correo": "carlos@example.com",
    "activo": False,
    "promedio": 3.8
}


alumnos = [
    estudiante1,
    estudiante2,
    estudiante3
]

# Imprimir el nombre de cada estudiante en la lista de estudiantes
print("Estudiantes registrados:")
print(alumnos)
```

## Condicionales y bucles

## Condicionales:  son estructuras de control de un flujo

if/elif/else

```python
nombre = "Ana"
promedio = 6.0

# Evaluar el desempeño del estudiante basado en su promedio
if promedio >= 7.0:
    print(f"{nombre} felicitaciones por su excelente desempeño con un promedio de {promedio}.")
else:
    print(f"{nombre} necesita mejorar su desempeño con un promedio de {promedio}.")

# Evaluar el desempeño del estudiante basado en su promedio con múltiples condiciones
if promedio >= 7.0:
    print(f"{nombre} felicitaciones por su excelente desempeño con un promedio de {promedio}.")
elif promedio >= 5.0:
    print(f"{nombre} tiene un desempeño aceptable con un promedio de {promedio}, pero puede mejorar.")
else:
    print(f"{nombre} necesita mejorar su desempeño con un promedio de {promedio}.")
```

## Bucles: son estructuras de iteracion

> for: recorre todo el elemento una cantidad de veces limitada

```python
learnes = [
    {
        "nombre": "Erika Restrepo",
        "promedio": 8.7
    },
    {
        "nombre": "Camilo Restrepo",
        "promedio": 7.5
    },
    {
        "nombre": "Carlos Mendoza",
        "promedio": 5.2
    }
]

# Evaluar el desempeño de cada estudiante en la lista de estudiantes
for learner in learnes:
   
    if promedio >= 7.0:
        print(f"{learner['nombre']} felicitaciones por su excelente desempeño con un promedio de {learner['promedio']}.")
    elif promedio >= 5.0:
        print(f"{learner['nombre']} tiene un desempeño aceptable con un promedio de {learner['promedio']}, pero puede mejorar.")
    else:
        print(f"{learner['nombre']} necesita mejorar su desempeño con un promedio de {learner['promedio']}.")
```
